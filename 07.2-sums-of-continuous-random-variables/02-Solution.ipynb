# Let $X$ and $Y$ be independent random variables defined on the space $\Omega$ with density functions $f_{X}$ and $f_{Y}$

# Suppose $Z=X+Y$

# Find the density $f_{Z}$ of $Z$ if:

## a) $f_{X} = f_{Y} = \left\{\begin{matrix}1/2 & -1\leq x \leq 1\\ 0 & otherwise\end{matrix}\right.$

## b) $f_{X} = f_{Y} = \left\{\begin{matrix}1/2 & 3\leq x \leq 5\\ 0 & otherwise\end{matrix}\right.$

## c) 

### $f_{X} = \left\{\begin{matrix}1/2 & -1\leq x \leq 1\\ 0 & otherwise\end{matrix}\right.$

### $f_{Y} = \left\{\begin{matrix}1/2 & 3\leq x \leq 5\\ 0 & otherwise\end{matrix}\right.$

_____

# a)

# First, we can visualize what the densities look like

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 10, 10
```

```python
X = np.arange(-1, 1.01, 0.01)
fX = 0.5*np.ones(len(X))
PX = np.cumsum(fX)/sum(fX)
plt.plot(X, fX, label = 'Density')
plt.plot(X, PX, label = 'Cumulative Distribution')
plt.xlim([-1,1])
plt.legend(loc = 'upper left')
plt.show()
```

**Let's take some samples and see a histogram**

```python
n = 10000
X_samples = np.random.uniform(-1, 1, size = n)
Y_samples = np.random.uniform(-1, 1, size = n)

Z_samples = X_samples + Y_samples
```

```python
pd.DataFrame(Z_samples).hist(bins = 100, normed = True)
```

### Recall: [Sum of Two Independent Uniform Random Variables](summary-of-chapter.ipynb#Sum-of-Two-Independent-Uniform-Random-Variables)

- This distribution looks similar to $f_{Z}$ in the example

# $f_{Z}(z) = \int_{-\infty}^{\infty}f_{X}(z-y)f_{Y}(y)dy$

# When $-1\leq y \leq 1$, this distribution is 1/2. Otherwise, it's zero

# We know $Z$ can take on any value in the interval $[-2, 2]$

# Let's split this into two regions: $[-2, 0]$ and $[0, 2]$

## 1) $[-2, 0]$

### For $f_{X}(z-y)$ and $f_{Y}(y)$ to be non-zero, we need $-1\leq y$ and $-1 \leq z-y \implies y \leq z+1$

### So $f_{Z}(z) = \int_{-1}^{z+1}\frac{1}{2}\frac{1}{2}dy = \frac{z+2}{4}$

## 2) $[0, 2]$

### For $f_{X}(z-y)$ and $f_{Y}(y)$ to be non-zero, we need $y \leq 1$ and $z-y \leq 1 \implies z-1 \leq y$

### So $f_{Z}(z) = \int^{1}_{z-1}\frac{1}{2}\frac{1}{2}dy = \frac{2-z}{4}$

# Therefore:

# $f_{Z}(z) = \left\{\begin{matrix} \frac{z+2}{4} & \text{ ; if }-2 \leq z\leq 0\\ \frac{2-z}{4} & \text{ ; if }0 \leq z \leq 2 \end{matrix}\right.$

## Let's plot this next to our histogram

```python
z1 = np.arange(-2, 0, 0.01)
fZ1 = (z1+2)/4

z2 = np.arange(0, 2, 0.01)
fZ2 = (2-z2)/4

plt.plot(z1, fZ1, lw = 3, color = 'r')
plt.plot(z2, fZ2, lw = 3, color = 'r')

plt.hist(Z_samples, bins = 100, normed = True, color = 'b')
plt.show()
```

**Pretty spot on**

# b)

```python
n = 10000
X_samples = np.random.uniform(3, 5, size = n)
Y_samples = np.random.uniform(3, 5, size = n)

Z_samples = X_samples + Y_samples
plt.hist(Z_samples, bins = 100, normed = True)
plt.show()
```

**Looks the same as a)**

# When $6\leq z \leq 8$, $3 \leq y$ and $3\leq z-y \implies y \leq z-3$ so

# $f_{Z}(z) = \int_{3}^{z-3}\frac{1}{4}dy = \frac{z-6}{4}$

# When $8\leq z \leq 10$, $5 \geq y$ and $5\geq z-y \implies y \geq z-5$ so

# $f_{Z}(z) = \int_{z-5}^{5}\frac{1}{4}dy = \frac{10-z}{4}$

# Therefore:

# $f_{Z}(z) = \left\{\begin{matrix} \frac{z-6}{4} & \text{ ; if }6 \leq z\leq 8\\ \frac{10-z}{4} & \text{ ; if }8 \leq z \leq 10 \end{matrix}\right.$

```python
z1 = np.arange(6, 8, 0.01)
fZ1 = (z1-6)/4

z2 = np.arange(8, 10, 0.01)
fZ2 = (10-z2)/4

plt.plot(z1, fZ1, lw = 3, color = 'r')
plt.plot(z2, fZ2, lw = 3, color = 'r')

plt.hist(Z_samples, bins = 100, normed = True, color = 'b')
plt.show()
```

# c)

```python
n = 10000
X_samples = np.random.uniform(-1, 1, size = n)
Y_samples = np.random.uniform(3, 5, size = n)

Z_samples = X_samples + Y_samples
plt.hist(Z_samples, bins = 100, normed = True)
plt.show()
```

# When $2\leq z \leq 4$, $3 \leq y$ and $-1\leq z-y \implies y \leq z+1$ so

# $f_{Z}(z) = \int_{3}^{z+1}\frac{1}{4}dy = \frac{z-2}{4}$

# When $4\leq z \leq 6$, $5 \geq y$ and $1\geq z-y \implies y \geq z-1$ so

# $f_{Z}(z) = \int_{z-1}^{5}\frac{1}{4}dy = \frac{6-z}{4}$

# Therefore:

# $f_{Z}(z) = \left\{\begin{matrix} \frac{z-2}{4} & \text{ ; if }2 \leq z\leq 4\\ \frac{6-z}{4} & \text{ ; if }4 \leq z \leq 6 \end{matrix}\right.$

```python
z1 = np.arange(2, 4, 0.01)
fZ1 = (z1-2)/4

z2 = np.arange(4, 6, 0.01)
fZ2 = (6-z2)/4

plt.plot(z1, fZ1, lw = 3, color = 'r')
plt.plot(z2, fZ2, lw = 3, color = 'r')

plt.hist(Z_samples, bins = 100, normed = True, color = 'b')
plt.show()
```
