# In Las Vegas, a roulette wheel has 38 slots numbered 0, 00, 1, 2, 3, ..., 36. The 0 and 00 slots are green and half the remaining 36 slots are red with the other half black.

# Write a program to find the total winnings for a player who makes 1000 bets on red

```python
import numpy as np
```

```python
def check(val):
    if (val==0)|(val==1):
        return -1
    elif val%2==1:
        return -1
    else:
        return 1
```

```python
check = np.vectorize(check)
```

```python
def spins(n_spins):
    rand_array = np.random.randint(38, size = n_spins)
    return np.cumsum(check(rand_array))
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 10, 7.5
```

```python
N_trials = 10
N_spins = 1000

for i in range(N_trials):
    x = range(N_spins)
    y = spins(N_spins)
    plt.plot(x,y)
plt.axhline(0, ls = 'dashed', color = 'r')
plt.show()
```
