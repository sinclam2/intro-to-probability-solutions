# Tversky and his colleagues studied the records of 48 Philadelphia 76ers basketball games in the 1980-1981 season to see if a player had times when he was hot and every shot went in, and other times when he was cold and could barely hit the backboard

# The players estimated that they were about 25% more likely to make a shot after a hit than after a miss.

# In fact, the opposite was true; they were 6% more likely to hit a shot after a miss.

# Tversky reports that the number of hot and cold streaks was about what one would expect by purely random effects.

# Assuming that a player has a 50/50 chance of making a shot and makes 20 shots/game, estimate by simulation the proportion of games in which the player will have a streak of 5 or more hits

```python
import numpy as np
import pandas as pd
```

```python
def sim():
    array = np.random.randint(0,2, size = 20)
    return array
```

```python
df = pd.DataFrame()

for i in range(10000):
    df[i] = sim()
```

```python
df
```

**Note**: the following function is deprecated in python 3

```python
((pd.rolling_mean(df, 5).max()==1)*1).sum()/float(10000)
```

# At least 25% of time there's a hot streak of 5 in a row
